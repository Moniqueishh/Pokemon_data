# Working with APIs

API stands for Application programming interface<br>
APIs are a way for applications to talk to eachother<br>
In python we use the requests package to handle the connection<br>
<small><strong>Note: It is requests with an s not request(this is a different package)</strong></small><br>
Most API's communicate with data in the form of JSON<br>
JSON stands for JavaScript Object Notation

```python
#import the packages we need:

import requests, json
```

```python
url = "http://ergast.com/api/f1/2008/5/driverStandings.json" #Need to have .json 
```

```python
# response object
response = requests.get(url)
print(response) #checking if the connection was successful 
```

```python
#check and make sure we got a successful response from the API
response.ok
```

```python
#view the response as a JSON
response.json()
```

```python
type(response.json())
data = response.json()
data
```

```python
# We only want the Driver Standings
x = data["MRData"]["StandingsTable"]["StandingsLists"][0]["DriverStandings"]
```

```python
def get_top(data):
    new_data = []
    for d in data:
        driver = {} # we are going to put each driver in their own dictionary 
        d_name = d["Driver"]["familyName"]
        
        driver[d_name] = {
            "first_name": d["Driver"]["givenName"],
            "last_name" : d["Driver"]["familyName"],
            "position" : d["position"],
            "nationality": d["Driver"]["nationality"],
            "team": d["Constructors"][0]["constructorId"],            
            
        }
        
        new_data.append(driver)
    return new_data
```

```python
get_top(x)
```

```python
def driver_results(year,round):
    url = f"http://ergast.com/api/f1/{year}/{round}/results.json" 
    response = requests.get(url)
    data = response.json()["MRData"]["RaceTable"]["Races"][0]["Results"]
#     print(data)
    for d in data:
        print(f"Finishing in po {d['position']} is {d['Driver']['driverId']}")
```

```python
driver_results(2013, 7)
```
